```python
import matplotlib.pyplot as plt
import os
```

```python
def read_loss(file_names, start_words='Train loss'):
#Train loss: 0.4434610605239868. Valid loss: 0.45140939950942993
    train_loss = []
    valid_loss = []
    for file_name in file_names:
        with open(file_name, 'r') as f:
            for row in f:
                if row.startswith(start_words):
                    train_loss.append(float(row.split(": ")[1][:-len(". Valid loss:")]))
                    valid_loss.append(float(row.split(": ")[2]))
    return train_loss, valid_loss
```

```python
files=["../matey-12K-baseline-N8-3110401.out",  "../matey-12K-baseline-N8-3114871.out", "../matey-12K-baseline-N8-3116247.out"]
train_loss, valid_loss=read_loss(files)
fig, axs = plt.subplots(1, 1, figsize=(12, 12))
lwd=2.0
axs.plot(train_loss, linewidth=lwd, label="Train")
axs.plot(valid_loss, linewidth=lwd, label="Valid")
axs.legend()
axs.set_xlabel("Epoch")
axs.set_ylabel("Loss")
```
